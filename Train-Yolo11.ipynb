```python
import torch
import ultralytics
from ultralytics import YOLO

ultralytics.checks()
print(f"Cuda: {torch.cuda.is_available()}")
```

```python

dataset_file = "Data/TinyYoloMulticlass/dataset.yaml" # make sure 'path' in yaml file matches the actual data path

# pretrained model
pretrained_model = "yolo11n-seg.pt"
device = "cuda" # or "cpu"
img_size = 512
epochs = 20
batch_size = 8

model = YOLO(pretrained_model, task="segment")   # 

model.train(
    data=dataset_file,
    imgsz=img_size,
    epochs=epochs,
    batch=batch_size,
    device=device,
)
```
